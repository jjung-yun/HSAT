```python
#  
n,k=map(int,input().split())
count=0
while n>1:
    if n%k==0:
        n/=k
    else:
        n-=1
    count+=1
print(count)
```

```python
#  
# N, K    
n, k = map(int, input().split())

result = 0

while True:
    # N K      1 
    target = (n // k) * k
    result += (n - target)
    n = target
    # N K   (     )  
    if n < k:
        break
    # K 
    result += 1
    n //= k

#     1 
result += (n - 1)
print(result)
```
